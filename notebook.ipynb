````python
# Pasar todos los nombres de columnas a minúsculas
games.columns = games.columns.str.lower()

# Mostramos los nombres nuevos para confirmar
print("Nombres de columnas después de pasar a minúsculas:")
print(games.columns.tolist())
print()
````

#Porque con estos cambios, el dataset está mucho más limpio y listo para análisis:

# year_of_release → entero lógico y compatible con NaN
# user_score → numérico real → permite todos los cálculos estadísticos necesarios en pasos posteriores (correlación con ventas, comparación entre géneros, etc.)


#Decidí eliminar únicamente las 2 filas que no tienen nombre ni género, ya que sin esta información básica no es posible analizar esos registros y su eliminación tiene un impacto mínimo en el dataset completo.
#En todas las demás columnas con valores ausentes (year_of_release, critic_score, user_score y rating) opté por **dejar los valores como NaN** y no rellenarlos con ninguna estimación (ni media, ni mediana, ni 0, ni ningún otro valor).
#La razón principal es que imputar estos datos introduciría sesgo artificial y distorsionaría los resultados posteriores, especialmente en análisis de correlación, promedios por plataforma o género y comparaciones entre regiones.  
#Los porcentajes de ausentes son muy altos en critic_score (~51%), user_score (~55%) y rating (~40%), por lo que cualquier relleno sería altamente especulativo y poco confiable. Mantener los NaN es la opción más transparente y correcta estadísticamente, ya que pandas los ignora automáticamente en cálculos relevantes.

¿Por qué crees que los valores están ausentes? 
 Los valores ausentes en el dataset tienen explicaciones lógicas relacionadas con la naturaleza de la industria de los videojuegos y la forma en que se recopilaron los datos:

 **name y genre** (2 casos): muy probablemente errores de captura o registros incompletos en la fuente original.  
 **year_of_release** (269 casos): puede deberse a juegos muy antiguos sin fecha exacta documentada, reediciones o ports sin año claro, o simplemente errores en bases de datos históricas.  
 **critic_score** (~51%): muchos juegos, especialmente los antiguos, los independientes, los de nicho o los lanzados solo en mercados locales (como Japón), nunca fueron reseñados por críticos profesionales.  
 **user_score** (~55%): hay juegos con muy pocas o ninguna reseña de usuarios (juegos poco populares, antiguos —antes de existir plataformas masivas como Metacritic—, o lanzamientos muy recientes con calificación aún pendiente).  
 **rating** (ESRB) (~40%): no todos los juegos reciben clasificación ESRB, sobre todo los que se lanzaron exclusivamente fuera de Norteamérica (Japón, Europa, etc.), juegos indie o títulos de bajo perfil que no pasaron por el proceso de clasificación.

Manejo de la abreviatura "tbd" en user_score

#La abreviatura **"tbd"** significa "to be determined" (aún por determinar / pendiente de determinar). Aparece cuando un juego no tiene aún una puntuación de usuarios calculada, normalmente porque tiene muy pocas reseñas o es un lanzamiento muy reciente.

#En el dataset encontramos **2,424** valores exactamente iguales a "tbd".

#**Decisión tomada:** convertir todos los "tbd" a NaN utilizando `pd.to_numeric(..., errors='coerce')`.

#**Justificación:**  
#"tbd" no representa una calificación real (ni baja, ni media, ni alta), sino simplemente la ausencia de dato suficiente. Tratar "tbd" como 0, 5 o cualquier otro número sería inventar información y sesgaría gravemente los promedios, correlaciones y cualquier análisis posterior. La forma más correcta y profesional es tratarlo como **dato faltante (NaN)**, igual que los demás valores que ya no estaban presentes.

<div class="alert alert-block alert-success">
<b>Comentario de Revisor          </b> <a class="tocSkip"></a>

Buen trabajo en el tratamiento de los valores faltantes. En aquellas columnas con una alta proporción de datos ausentes, es buena idea optar por mantener los nulos o reemplazarlos por un valor que permita identificarlos claramente. En particular, para critic_score y user_score, la cantidad de información disponible no es suficiente para realizar una imputación confiable. Completar estos campos con la media o la mediana podría distorsionar significativamente los resultados, introduciendo un sesgo importante en el análisis, por lo que la decisión tomada resulta adecuada y bien justificada.



</div>

Construye una distribución basada en los datos de cada año

````python
import matplotlib.pyplot as plt

# Seleccionamos solo las top 10
top_plataformas = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS', 'GBA', 'PS4', 'PSP', 'PC']

# Tabla de ventas por año (solo últimos 12 años para no saturar)
pivot_ventas = games.pivot_table(
    index='year_of_release',
    columns='platform',
    values='total_sales',
    aggfunc='sum',
    fill_value=0
)[top_plataformas]

print("Ventas anuales - Top 10 plataformas (años recientes):")
display(pivot_ventas.loc[2005:2016].round(2))  # Ajusta el rango si quieres ver más años

# Gráfico de líneas (muy claro para ver variación)
pivot_ventas.loc[2005:2016].plot(figsize=(12, 6), linewidth=2, marker='o')
plt.title('Evolución de ventas anuales - Top 10 plataformas históricas (2005-2016)')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Ventas totales (millones)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

 Pruebas de hipótesis

**Hipótesis 1:**  
Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

**Hipótesis nula (H₀):**  
μ_XOne = μ_PC  (las medias de user_score son iguales)

**Hipótesis alternativa (H₁):**  
μ_XOne ≠ μ_PC  (las medias son diferentes)

**Nivel de significancia (alfa):**  
0.05 (umbral estándar)

**Prueba:** t-test de dos muestras independientes (asumiendo varianzas no necesariamente iguales)

````python
# Importar librería para prueba t
from scipy.stats import ttest_ind

# 1. Filtrar datos relevantes (2013-2016) para Xbox One y PC
xone_scores = games_relevantes[
    (games_relevantes['platform'] == 'XOne') & 
    (games_relevantes['user_score'].notna())
]['user_score']

pc_scores = games_relevantes[
    (games_relevantes['platform'] == 'PC') & 
    (games_relevantes['user_score'].notna())
]['user_score']

# 2. Ver número de observaciones
print("Número de calificaciones Xbox One:", len(xone_scores))
print("Número de calificaciones PC:", len(pc_scores))

# 3. Medias y desviaciones estándar
print(f"Media Xbox One: {xone_scores.mean():.2f}")
print(f"Media PC: {pc_scores.mean():.2f}")

# 4. Prueba t de Student (independientes, varianzas no iguales)
t_stat, p_value = ttest_ind(xone_scores, pc_scores, equal_var=False)

print(f"\nEstadístico t: {t_stat:.3f}")
print(f"p-value: {p_value:.5f}")

# 5. Decisión
if p_value < 0.05:
    print("Rechazamos H₀ → Las calificaciones promedio son DIFERENTES")
else:
    print("No rechazamos H₀ → No hay evidencia suficiente para decir que son diferentes")
````

 Pruebas de hipótesis – Hipótesis 2

 Hipótesis:
 H₀ (nula): Las calificaciones promedio de usuarios para los géneros Acción y Deportes son IGUALES
 H₁ (alternativa): Las calificaciones promedio de usuarios para Acción y Deportes son DIFERENTES

 Nivel de significancia (alfa) elegido: 0.05 (estándar, suficiente para rechazar H₀ con confianza razonable)

 Criterio utilizado: prueba t de Student para dos muestras independientes (ttest_ind)
 Razón: 
 - Comparación de medias de dos grupos independientes (Acción vs Deportes)
 - Muestras grandes (>30 por grupo) → t-test es robusto
 - Varianzas no necesariamente iguales → equal_var=False
 - Decisión: si p-value < 0.05 → rechazamos H₀ (diferencia significativa)

````python
# Importar librería para prueba t (si no está importada)
from scipy.stats import ttest_ind

# 1. Filtrar datos relevantes (2013-2016) para géneros Acción y Deportes
action_scores = games_relevantes[
    (games_relevantes['genre'] == 'Action') & 
    (games_relevantes['user_score'].notna())
]['user_score']

sports_scores = games_relevantes[
    (games_relevantes['genre'] == 'Sports') & 
    (games_relevantes['user_score'].notna())
]['user_score']

# 2. Ver número de observaciones y medias
print("Número de calificaciones Acción:", len(action_scores))
print("Número de calificaciones Deportes:", len(sports_scores))
print(f"Media Acción: {action_scores.mean():.2f}")
print(f"Media Deportes: {sports_scores.mean():.2f}")

# 3. Prueba t de Student (varianzas no iguales)
t_stat, p_value = ttest_ind(action_scores, sports_scores, equal_var=False)

print(f"\nEstadístico t: {t_stat:.3f}")
print(f"p-value: {p_value:.5f}")

# 4. Decisión
if p_value < 0.05:
    print("Rechazamos H₀ → Las calificaciones promedio son DIFERENTES")
else:
    print("No rechazamos H₀ → No hay evidencia suficiente para decir que son diferentes")
````
